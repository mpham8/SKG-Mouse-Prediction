```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
taxa_df = pd.read_csv('taxa.csv')
otu_df = pd.read_csv('otu.csv')
sample_data_df = pd.read_csv('sample_data.csv')

otu_df
```

```python
otu_df.rename(columns={ otu_df.columns[0]: 'id' }, inplace = True)
otu_delete_lst = []

otu_lst = otu_df.iloc[:,0].tolist()
for otu in otu_lst:
    otu_row = otu_df.loc[otu_df['id'] == otu]
    otu_values_lst = otu_row.iloc[0, 1:].tolist()
    for value in otu_values_lst:
        if value >= 0.1:
            break
    else:
        otu_delete_lst.append(otu)
        
otu_df = otu_df[~otu_df['id'].isin(otu_delete_lst)]
otu_df
```

```python
targets = []
samples = otu_df.columns[1:]
for sample in samples:
    target = sample_data_df.loc[sample_data_df['SampleLabel'] == sample, 'Genotype'].iloc[0]
    print(target)
    
    #SKGNak -> 0
    #WT -> 1
    #SKGBoyJ ->2
    if 'SKG' in target:
        label = 0
    elif 'WT' in target:
        label = 1
    # elif 'SKGBoyJ' in target:
    #     label = 2
    targets.append(label)

    
targets
```

```python
all_zeros = []
for i in range (len(otu_df)):
    x = []
    zeros = 0
    lst = otu_df.iloc[i,1:].tolist()
    for j in range(len(lst)):
        if targets[j] == 0:
            x.append(lst[j])
        if lst[j] == 0:
            zeros += 1
    p_zeros = zeros/len(lst)
    all_zeros.append(p_zeros)
    print(p_zeros)
    print(np.sum(lst)/len(lst))
    print(np.std(lst))
    plt.hist(x, 20)
    plt.show() 

print('zero plot')
plt.hist(all_zeros, 30)
plt.show() 
print(np.sum(all_zeros)/len(otu_df))
```
